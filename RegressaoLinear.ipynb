```python
import matplotlib.pyplot as plt
import numpy as np
```

Valores durao da erupo x por intervalo at a prxima erupo y



```python
#Valores de x e y vetorizados
x = [1.80, 1.82, 1.90, 1.93, 1.98, 2.05, 2.13, 2.30, 2.37, 2.82, 3.13, 3.27, 3.65, 3.78, 3.83, 3.88, 4.10, 4.27, 4.30, 4.43, 4.47, 4.53, 4.55, 4.60, 4.63]
y = [56, 58, 62, 56, 57, 57, 60, 57, 61, 73, 76, 77, 77, 79, 85, 80, 89, 90, 89, 89, 86, 89, 86, 92, 91]

#print(len(x))
#print(len(y))
#Ambos vetores possuem o mesmo tamanho

#Calculos necessrios para determinar a reta de regresso linear y = a + bx
n = len(x)

somatorio_x = 0
for i in x:
    somatorio_x += i

somatorio_y = 0
for i in y:
    somatorio_y += i

somatorio_xy = 0
for i in range(n):
    valor_x = x[i]
    valor_y = y[i]

    somatorio_xy += valor_x * valor_y

quadrado_x = 0
for i in x:
    quadrado_x += pow(i, 2)

media_x = somatorio_x / n
media_y = somatorio_y / n

print(somatorio_x, somatorio_y, somatorio_xy, quadrado_x, media_x, media_y)
```

```python
#Calculo dos valores das variveis a (b0) e b(b1)
b = ((n * somatorio_xy) - (somatorio_x * somatorio_y)) / ((n * quadrado_x) - (pow(somatorio_x, 2)))

a = media_y - (b * media_x)
print(a, b)
```

```python
# Plotando o grfico
x_np = np.array(x)
y_pred = a + b * x_np

plt.figure(figsize=(10,6))

#Dados Originais
plt.scatter(x, y, label='Dados Originais (Durao vs. Intervalo)', color='blue', alpha=0.7)

#Regresso Linear
plt.plot(x, y_pred, label=f'Reta de Regresso: y = {a:.2f} + {b:.2f}x', color='red', linestyle='--')

plt.title('Grfico de Disperso com Reta de Regresso Linear')
plt.xlabel('Durao da Erupo (minutos)')
plt.ylabel('Intervalo at a Prxima Erupo (minutos)')
plt.legend()
plt.grid(True)
plt.show()
```
